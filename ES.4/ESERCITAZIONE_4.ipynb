##### _SISTEMA ALLO STATO SOLIDO_

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_path):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(os.path.join(os.getcwd(), file_path), 'r') as file:
            # Ignora la prima riga
            next(file)
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore
                dati_riga = riga.strip().split()
                
                # Converte i dati in float e aggiunge alla lista dati
                dati_float = [float(dato) for dato in dati_riga[2:4]]  # Terza e quarta colonna
                dati.append(dati_float)
    except FileNotFoundError:
        print(f"Errore: il file {file_path} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy




# Dati per il primo grafico
file_dati_total= "OUTPUT_solido/total_energy.dat"
dati_total = takeData(file_dati_total)
x_values_total = dati_total[:, 0]  # Valori
y_errors_total = dati_total[:, 1]  # Errori


# Dati per il secondo grafico
file_dati_pot = "OUTPUT_solido/potential_energy.dat"
dati_pot = takeData(file_dati_pot)
x_values_pot = dati_pot[:, 0]  # Valori
y_errors_pot = dati_pot[:, 1]  # Errori

# Dati per il terzo grafico
file_dati_kin = "OUTPUT_solido/kinetic_energy.dat"
dati_kin = takeData(file_dati_kin)
x_values_kin = dati_kin[:, 0]  # Valori
y_errors_kin = dati_kin[:, 1]  # Errori

# Dati per il quarto grafico
file_dati_temp = "OUTPUT_solido/temperature.dat"
dati_temp= takeData(file_dati_temp)
x_values_temp = dati_temp[:, 0]  # Valori
y_errors_temp = dati_temp[:, 1]  # Errori

# Dati per il quinto grafico
file_dati_press = "OUTPUT_solido/pressure.dat"
dati_press= takeData(file_dati_press)
x_values_press = dati_press[:, 0]  # Valori
y_errors_press = dati_press[:, 1]  # Errori

# Parametri comuni
#50 000 lanci in totale
M = 50   # Numero di blocchi
N = 108 #numero di particelle
L = 1

x = np.arange(0, M)

# Crea i subplot affiancati
fig, axs = plt.subplots(3, 2, figsize=(14, 14))
# Aggiungi spazio tra i grafici
plt.subplots_adjust(hspace=0.5, wspace=0.5)


# Primo grafico
axs[0, 0].errorbar(x, x_values_total, yerr=y_errors_total)
axs[0, 0].set_xlabel('#Blocchi')
axs[0, 0].set_ylabel('<E>/N')
axs[0, 0].set_title('Energia media per particella allo stato solido')
axs[0, 0].grid(True)
axs[0, 0].set_xticks(np.arange(0, M+1, 5))



# Secondo grafico
axs[0, 1].errorbar(x, x_values_pot, yerr=y_errors_pot)
axs[0, 1].set_xlabel('#Blocchi')
axs[0, 1].set_ylabel('<U>/N')
axs[0, 1].set_title('Energia potenziale media per particella allo stato solido')
axs[0, 1].grid(True)
axs[0, 1].set_xticks(np.arange(0, M+1, 5))


# Terzo grafico
axs[1, 0].errorbar(x, x_values_kin, yerr=y_errors_kin)
axs[1, 0].set_xlabel('#Blocchi')
axs[1, 0].set_ylabel('<K>/N')
axs[1, 0].set_title('Energia cinetica media per particella allo stato solido')
axs[1, 0].grid(True)
axs[1, 0].set_xticks(np.arange(0, M+1, 5))

# Quarto grafico
axs[1, 1].errorbar(x, x_values_temp, yerr=y_errors_temp)
axs[1, 1].set_xlabel('#Blocchi')
axs[1, 1].set_ylabel('T')
axs[1, 1].set_title('Temperatura media per sistema allo stato solido')
axs[1, 1].grid(True)
axs[1, 1].set_xticks(np.arange(0, M+1, 5))

# Quinto grafico
axs[2, 0].errorbar(x, x_values_press, yerr=y_errors_press)
axs[2, 0].set_xlabel('#Blocchi')
axs[2, 0].set_ylabel('P')
axs[2, 0].set_title('Pressione media per sistema allo stato solido')
axs[2, 0].grid(True)
axs[2, 0].set_xticks(np.arange(0, M+1, 5))



axs[2, 1].axis('off')
plt.show()


```
