#### ESERCIZIO 8.2

Il _main.cpp_ si sviluppa in tre sezioni:  
- nella prima viene stimato _step_size_ (successivamente commentato);
- nella seconda si svolge il _Simulated Annealing_;
- nella terza si calcola il valor medio <H> per i parametri che minimizzano il sistema, 
e si campiona la funzione d'onda associata.  

I valori di $\mu$ e $\sigma$ trovati sono:
$\mu = 0.763662$,  $\sigma = 0.653607$

La _fase di raffreddamento_ è stata implementata usando la legge $T = T_{0} \cdot a^k $,  
dove $k$ è l'iterazione corrente, $a$ un parametro molto vicino a 1: $a = 0.999$.



```python
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_path):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(file_path, 'r') as file:
            for riga in file:
                # Separa i dati utilizzando qualsiasi quantità di spazio o tab come delimitatore
                dati_riga = riga.strip().split()
                
                # Converte i dati in float e aggiunge alla lista dati
                if len(dati_riga) >= 2:  # Ensure there are at least 2 elements
                    dati_float = [float(dato) for dato in dati_riga[:2]]
                    dati.append(dati_float)
                
    except FileNotFoundError:
        print(f"Errore: il file {file_path} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy
```

```python
#Grafico <H> in funzione degli step di SA
dati_SA = takeData("SA.dat") 
x_values_SA = dati_SA[:, 0]  
y_errors_SA = dati_SA[:, 1]

L = len(x_values_SA)
steps = np.arange(0,L, 1)

temp= np.loadtxt("temp.dat")


# Creazione della figura con due subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Primo grafico
ax1.errorbar(steps, x_values_SA, yerr=y_errors_SA, ecolor='red')
ax1.set_xlabel('#Iterazioni S.A.')
ax1.set_ylabel('<H>')
ax1.set_title('Andamento di <H>')
ax1.grid(True)

# Secondo grafico
ax2.plot(steps, temp, label = "$T = T_{0} \cdot a^k $" )
ax2.set_xlabel('#Iterazioni S.A.')
ax2.set_ylabel('Temp')
ax2.set_title('Andamento della temperatura')
ax2.grid(True)
ax2.legend()

# Mostra i grafici
plt.tight_layout()
plt.show()

```

```python
#Grafico della media a blocchi per il <H>_min
dati_min = takeData("energie.dat") 
x_values_min = dati_min[:, 0]  
y_errors_min = dati_min[:, 1]

L = len(x_values_min)
block = np.arange(0,L, 1)

plt.errorbar(block, x_values_min, yerr=y_errors_min, ecolor='red')
plt.xlabel('#Blocchi')
plt.ylabel('<H>')
plt.title('Media a Blocchi per <H>$_{min}$')
plt.grid(True)
```

```python
data = np.loadtxt('psi.dat')
def wave(x):
   return (np.exp((-(x-mu)**2)/(2*sigma**2)) + np.exp((-( x + mu)**2)/(2*sigma**2)))

#modulus squared norm calculated using mathematica 
def norm(mu,sigma):
    return 2*(1+np.exp(-mu**2/sigma**2)) *np.sqrt(np.pi)*abs(sigma)

def wave2(x,mu,sigma):
   return wave(x)**2/norm(mu,sigma)

mu = 0.763662
sigma = 0.653607
x = np.linspace(-5,5,1000)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=200, edgecolor='black',  density=True)  
plt.plot(x,wave2(x,mu,sigma),label='analytical')
plt.xlim(-2,2)
plt.xlabel('Valore')
plt.ylabel('Frequenza')
plt.title('Istogramma distribuzione di probabilità dello stato fondamentale')
plt.grid(True)
plt.show()
```
