Nell'_header file random.h_ è stata aggiunta la classe _GBM_, per implementare il moto _browniano_ geometrico.   
La classe, oltre al costruttore, contiene due metodi: 
- _Valuta_  per campionare direttamente il prezzo finale dell'attività $ \ S(T)$;
- _Valuta_discr_  per campionare il percorso discretizzato del  $GBM(r,\sigma^2)$ del prezzo dell'attività, dividendo $[0,T]$ in $10^2$ intervalli di tempo:  
     $S(0)$, $S(t_1)$, $S(t_2) \ $, ... fino a $ \ S(t_{100}=T=1)$.
<br><br>

Sono state svolte $10^5$ simulazioni per ogni tipo di campionamento e sono stati stimati i prezzi, con l'errore, per le opzioni europee di _call_ e di _put_, applicando il metodo della _media a blocchi_.




```python
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_nome):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(file_nome, 'r') as file:
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore (puoi modificare in base al formato)
                dati_riga = riga.strip().split()
                
                # Converte i dati in float e aggiunge alla lista dati
                dati_float = [float(dato) for dato in dati_riga]
                dati.append(dati_float)
    except FileNotFoundError:
        print(f"Errore: il file {file_nome} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy

# Dati per il primo grafico
file_dati_call = "Call.out"
dati_call = takeData(file_dati_call)
x_values_call = dati_call[:, 0]  # Valori
y_errors_call = dati_call[:, 1]  # Errori

# Dati per il secondo grafico
file_dati_put = "Put.out"
dati_put = takeData(file_dati_put)
x_values_put = dati_put[:, 0]  # Valori
y_errors_put = dati_put[:, 1]  # Errori

# Dati per il terzo grafico
file_dati_call_discr = "Call_discr.out"
dati_call_discr = takeData(file_dati_call_discr)
x_values_call_discr = dati_call_discr[:, 0]  # Valori
y_errors_call_discr = dati_call_discr[:, 1]  # Errori

# Dati per il quarto grafico
file_dati_put_discr = "Put_discr.out"
dati_put_discr = takeData(file_dati_put_discr)
x_values_put_discr = dati_put_discr[:, 0]  # Valori
y_errors_put_discr = dati_put_discr[:, 1]  # Errori

# Parametri comuni

N = 100   # Numero di blocchi


x = np.arange(0, N, 1)

# Crea i subplot affiancati
fig, axs = plt.subplots(2, 2, figsize=(14, 14))

# Primo grafico
axs[0, 0].errorbar(x, x_values_call, yerr=y_errors_call)
axs[0, 0].axhline(y=14.975790778311286, color='r', linestyle='--')
axs[0, 0].set_xlabel('#Blocchi')
axs[0, 0].set_ylabel('C[ S(t),t ]')
axs[0, 0].set_title('Call Option - diretto')
axs[0, 0].grid(True)


# Secondo grafico
axs[0, 1].errorbar(x, x_values_put, yerr=y_errors_put)
axs[0, 1].axhline(y=5.4595325819072364, color='r', linestyle='--')
axs[0, 1].set_xlabel('#Blocchi')
axs[0, 1].set_ylabel('P[ S(t),t ]')
axs[0, 1].set_title('Put Option - diretto')
axs[0, 1].grid(True)


# Terzo grafico
axs[1, 0].errorbar(x, x_values_call_discr, yerr=y_errors_call_discr)
axs[1, 0].axhline(y=14.975790778311286, color='r', linestyle='--')
axs[1, 0].set_xlabel('#Blocchi')
axs[1, 0].set_ylabel('C[ S(t),t ]')
axs[1, 0].set_title('Call Option - discreto')
axs[1, 0].grid(True)


# Quarto grafico
axs[1, 1].errorbar(x, x_values_put_discr, yerr=y_errors_put_discr)
axs[1, 1].axhline(y=5.459, color='r', linestyle='--')
axs[1, 1].set_xlabel('#Blocchi')
axs[1, 1].set_ylabel('P[ S(t),t]')
axs[1, 1].set_title('Put Option - discreto')
axs[1, 1].grid(True)



plt.show()
```
