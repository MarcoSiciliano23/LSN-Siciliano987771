#### ESERCIZIO 9.1

Nell' _header file ingiro.h_ sono state implementate tre classi per trattare _il problema del commesso viaggiatore_, utilizzando un algoritmo genetico. 

La classe _City_ è stata creata per accedere alle coordinate dei punti con una sintassi più semplice; di fatto, il costruttore crea un _**std::vector**_ bidimensionale.  
La classe _Mappa_ pone una serie di _City_ in un _**std::vector**_, il costruttore imposta le coordinate di ogni città tramite i metodi della classe _City_.  
In questa classe è presente il metodo che calcola la distanza totale del percorso, ricevendo in input un _**std::vector<\int>**_ che rappresenta l'ordine in cui visitare le città. Il metodo utilizza questi indici per calcolare la distanza euclidea tra le città consecutive nel percorso, inclusa la distanza tra l'ultima città e la prima per chiudere il ciclo.

La classe _Genetica_ raggruppa tutti i metodi necessarî per effettuare le mutazioni e il _crossover_. Il costruttore di questa classe riempie un _**std::Vector<\int>**_ con interi da 0 a _ncity_, cosicché il primo percorso associ ogni intero all'indice della _Mappa_ (_**std::Vector<\City>**_), e dunque ogni città sia univocamente associata a un solo intero.  
Inoltre, il costruttore riceve in input una _**string**_ che specifica se costruire una mappa sulla circonferenza unitaria o dentro il quadrato unitario.  

Il metodo _Evoluzione_ è il cuore dell'algoritmo: mette insieme tutti i metodi della classe per far evolvere il sistema, restituendo il miglior percorso.  
Per terminare l'evoluzione, anziché impostare un numero fissato di generazioni, viene utilizzato un contatore che monitora l'assenza di aggiornamenti nella soluzione migliore: se dopo 25<sup>*</sup> generazioni la soluzione rimane la stessa, l'algoritmo si interrompe. 


<sup>*</sub><sub><sup> Stima "empirica" basata sulla complessità di questo problema </sup></sub>
##### Specifiche:
No. individui nella popolazione:  $1000$  
Prob. _crossover_:  $90\%$  
Prob. _shift_:  $10\%$  
Prob. _m-permutazioni contigue_:  $10\%$  
Prob. _permutazione semplice_:  $5\%$  
Prob. _inversione_:  $7.5\%$ 


```python
import matplotlib.pyplot as plt
import numpy as np

# Carica i dati
def takeData(file_path):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(file_path, 'r') as file:
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore
                dati_riga = riga.strip().split()
                
                # Converte i primi due dati in float e aggiunge alla lista dati
                if len(dati_riga) >= 2:  # Assicura che ci siano almeno 2 elementi
                    dati_float = [float(dato) for dato in dati_riga[:2]]
                    dati.append(dati_float)
                
    except FileNotFoundError:
        print(f"Errore: il file {file_path} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    # Converte la lista di liste in un array NumPy
    return np.array(dati)

# Carica i dati per il cerchio
dati_cerchio = takeData('length_cerchio.dat')
gen_cerc = dati_cerchio[:, 0]  # Prima colonna
dist_cerc = dati_cerchio[:, 1]  # Seconda colonna

# Carica i dati per il quadrato
dati_quadrato = takeData('length_quad.dat')
gen_quad = dati_quadrato[:, 0]  # Prima colonna
dist_quad = dati_quadrato[:, 1]  # Seconda colonna



# Crea una figura con due subplot affiancati
fig, axx = plt.subplots(1, 2, figsize=(12, 6), tight_layout=True)

# Grafico per il fitness medio per generazione per i dati del cerchio nel primo subplot
axx[0].plot(gen_cerc, dist_cerc, linewidth=0.8, marker=".")
axx[0].set_xlabel("#Generazione")
axx[0].set_ylabel(r"$\langle L^{(1)} \rangle$")
axx[0].set_title("Fitness medio per generazione sulla circonferenza unitaria")
axx[0].grid(True)

# Grafico per il fitness medio per generazione per i dati del quadrato nel secondo subplot
axx[1].plot(gen_quad, dist_quad, linewidth=0.8, marker=".")
axx[1].set_xlabel("#Generazione")
axx[1].set_ylabel(r"$\langle L^{(1)} \rangle$")
axx[1].set_title("Fitness medio per generazione nel quadrato unitario")
axx[1].grid(True)

# Mostra i grafici
plt.show()
```

##### PERCORSI MIGLIORI

```python
import matplotlib.pyplot as plt

# Carica il percorso
def carica_percorso(file):
    percorso = []
    with open(file, 'r') as f:
        for line in f:
            citta = int(line.strip())
            percorso.append(citta)
    return percorso

# Carica i dati per il cerchio
dati_cerchio = takeData('coord_cerc.dat')
x_cerc = dati_cerchio[:, 0]  # Prima colonna
y_cerc = dati_cerchio[:, 1]  # Seconda colonna

# Carica i dati per il quadrato
dati_quadrato = takeData('coord_quad.dat')
x_quad = dati_quadrato[:, 0]  # Prima colonna
y_quad = dati_quadrato[:, 1]  # Seconda colonna

# Carica il percorso
percorso_cerc = carica_percorso('percorso_cerc.dat')
percorso_quad = carica_percorso('percorso_quad.dat')

# Crea una figura con due subplot affiancati
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Grafico per le città su una circonferenza
axs[0].scatter(x_cerc, y_cerc)
axs[0].set_title('Percorso migliore per delle città sulla circonferenza unitaria')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

percorso_coordinate_cerc = [(x_cerc[i], y_cerc[i]) for i in percorso_cerc]
px_cerc, py_cerc = zip(*percorso_coordinate_cerc)
axs[0].plot(px_cerc, py_cerc, color='black', linewidth=1)
axs[0].plot([px_cerc[-1], px_cerc[0]], [py_cerc[-1], py_cerc[0]], color='black', linewidth=1)

# Grafico per le città in un quadrato
axs[1].scatter(x_quad, y_quad)
axs[1].set_title('Percorso migliore per delle città dentro il quadrato unitario')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

percorso_coordinate_quad = [(x_quad[i], y_quad[i]) for i in percorso_quad]
px_quad, py_quad = zip(*percorso_coordinate_quad)
axs[1].plot(px_quad, py_quad, color='black', linewidth=1)
axs[1].plot([px_quad[-1], px_quad[0]], [py_quad[-1], py_quad[0]], color='black', linewidth=1)

# Mostra i grafici
plt.tight_layout()
plt.show()
```
