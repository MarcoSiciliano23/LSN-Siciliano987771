#### ESERCIZIO 1.1
Attraverso il generatore di numeri casuali uniformemente distribuiti è stato stimato $\langle r \rangle = \int_0^1 r dr = 1/2$,  
confermando l'uniformità di tale generatore.
Successivamente è stata stimanta $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$.  
Infine, è stato simulato anche un test del $\chi^2$, per rigettare l'ipotesi secondo cui i numeri NON seguissero una distribuzione uniforme.

```python
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_nome):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(file_nome, 'r') as file:
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore (puoi modificare in base al formato)
                dati_riga = riga.strip().split()
                
                # Converte i dati in float e aggiunge alla lista dati
                dati_float = [float(dato) for dato in dati_riga]
                dati.append(dati_float)
    except FileNotFoundError:
        print(f"Errore: il file {file_nome} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy

# Primo grafico
file_dati1 = "Mean.out"
dati1 = takeData(file_dati1)
x_values1 = dati1[:, 0]  # Valori
y_errors1 = dati1[:, 1]   # Errori
             
N = 100                 # Numero di blocchi

x = np.arange(0, N, 1)

plt.errorbar(x, x_values1 - 0.5, yerr=y_errors1)
plt.xlabel('#Blocchi')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()

# Secondo grafico
file_dati2 = "Scarti.out"
dati2 = takeData(file_dati2)
x_values2 = dati2[:, 0]  # Valori
y_errors2 = dati2[:, 1]   # Errori
            
N = 100                 # Numero di blocchi


plt.errorbar(x, x_values2 - 1/12, yerr=y_errors2)
plt.xlabel('#Blocchi')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()

# Terzo grafico
# Leggi i dati da un file
def leggi_dati(nome_file):
    dati = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                dati.append(float(riga.strip()))
    except FileNotFoundError:
        print(f"Errore: il file {nome_file} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file {nome_file}: {e}")
    return dati

# Plotta l'istogramma dei dati
def plotta_istogramma(dati, titolo, xlabel, ylabel, bins=15):
    plt.hist(dati, bins=bins, color='blue', edgecolor='black', alpha=0.7)
    plt.title(titolo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# File contenente i dati
file_dati3 = "Chi2_j.out"

# Leggi i dati dal file
dati3 = leggi_dati(file_dati3)

# Plotta l'istogramma dei dati
plotta_istogramma(dati3, "Istogramma dei dati", "Valore", "Frequenza")
```

#### ESERCIZIO 1.2

Alla classe _Random_ sono stati aggiunti i metodi relativi all'estrazione di numeri casuali secondo la distribuzione esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$, e la distribuzione di Cauchy-Lorentz $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$, attraverso l'inversione della cumulativa. 

Sono poi stati generati degli istogrammi con $10^4$ valori estratti secondo tali distribuzioni, per diversi valori di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$).

Si osservano le conseguenze del _teorema del limite centrale_ le distribuzioni con $\sigma$ definita convergono a una distribuzione gaussiana, mentre la distribuzione di Cauchy_Lorentz è _limite di sé stessa_, infatti non ha varianza finita.

```python

def plotta_istogramma(ax, dati, titolo, xlabel, ylabel, intervallo):
    ax.hist(dati, bins=100, range=intervallo, color='blue', edgecolor='black', alpha=0.7)
    ax.set_title(titolo)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)


# Leggi e plotta gli istogrammi a due a due
tries_array = np.array([1, 2, 10, 100])
intervallo = (-20, 20)

for k in range(0, 4, 2):  # Incremento di 2 per scorrere a due a due
    i1, i2 = tries_array[k], tries_array[k+1]
    
    nome_file1_lor = f'Lor{i1}.out'
    dati1_lor = leggi_dati(nome_file1_lor)
    nome_file1_unif = f'Unif{i1}.out'
    dati1_unif = leggi_dati(nome_file1_unif)
    nome_file1_exp = f'Exp{i1}.out'
    dati1_exp = leggi_dati(nome_file1_exp)
    
    nome_file2_lor = f'Lor{i2}.out'
    dati2_lor = leggi_dati(nome_file2_lor)
    nome_file2_unif = f'Unif{i2}.out'
    dati2_unif = leggi_dati(nome_file2_unif)
    nome_file2_exp = f'Exp{i2}.out'
    dati2_exp = leggi_dati(nome_file2_exp)
    
    if (dati1_lor is not None and dati1_unif is not None and dati1_exp is not None and
        dati2_lor is not None and dati2_unif is not None and dati2_exp is not None):
        
        fig, axs = plt.subplots(2, 3, figsize=(15, 8))  # Crea una griglia di subplot 2x3
        
        plotta_istogramma(axs[0, 0], dati1_lor, f'Lorentz No. lanci {i1}', 'Valore', 'Frequenza', intervallo)
        plotta_istogramma(axs[0, 1], dati1_unif, f'Uniforme No. lanci {i1}', 'Valore', 'Frequenza', None)
        plotta_istogramma(axs[0, 2], dati1_exp, f'Esponenziale No. lanci {i1}', 'Valore', 'Frequenza', None)
        
        plotta_istogramma(axs[1, 0], dati2_lor, f'Lorentz No. lanci {i2}', 'Valore', 'Frequenza', intervallo)
        plotta_istogramma(axs[1, 1], dati2_unif, f'Uniforme No. lanci {i2}', 'Valore', 'Frequenza', None)
        plotta_istogramma(axs[1, 2], dati2_exp, f'Esponenziale No. lanci {i2}', 'Valore', 'Frequenza', None)
        
        plt.tight_layout()
        plt.show()
```

#### ESERCIZIO 3

Per simulare l'esperimento di Buffon si è seguito il seguente approccio:  

 - viene generato un numero casuale tra 0 e $d$ (la distanza tra le linee del piano), questo rappresenta il centro dell'ago $x_c$;
 - vengono generati quattro numeri casuali $(x_1, y_1) \ (x_2, y_2)$, da cui si estrae il coefficiente angolare della retta passante per i due punti, così viene generato casualmente l'angolo d'inclinazione rispetto alle linee del piano;
 - dal centro si estende un segmento pari a $L/2$ a destra e a sinistra.

Così è stato modellizato il lancio dell'ago. Se a partire dal centro $x_c$ l'incremento di $+ L/2$ è maggiore di $d$ oppure il decremnto di $-L/2$ è minore di 0, allora l'ago ha colpito una delle linee parallele del piano; dunque viene aggiornato il contatore $N_{hit}$.

Si calcola il valore di $\pi$ dalla formula $$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
Viene calcolata l'incertezza sfruttando il metodo della _media a blocchi_.  
Nello specifico, sono stati effettuati $10^6$ lanci totali e $10^4$ lanci per blocco ($N_{thr}$).
