#### ESERCIZIO 7.2


```python
#PUNTO PRIMO

import os
import numpy as np
import matplotlib.pyplot as plt

# Funzione per prendere i dati dal file
def takeData(file_path):
    dati = []
    try:
        with open(os.path.join(os.getcwd(), file_path), 'r') as file:
            next(file)  # Salta l'intestazione
            for riga in file:
                dati_riga = riga.strip().split()
                dato = float(dati_riga[1])
                dati.append(dato)
    except FileNotFoundError:
        print(f"Errore: il file {file_path} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)

directory = str(os.getcwd())

# Percorsi ai file dei dati
file_dati_SOLIDO = os.path.join(directory, "Monte_Carlo/OUTPUT_solido/potential_energy_50.dat")
file_dati_LIQUIDO = os.path.join(directory, "Monte_Carlo/OUTPUT_liquido/potential_energy_50.dat")
file_dati_GAS = os.path.join(directory, "Monte_Carlo/OUTPUT_gas/potential_energy_50.dat")

# Lettura dei dati
dati_SOLIDO = takeData(file_dati_SOLIDO)
dati_LIQUIDO = takeData(file_dati_LIQUIDO)
dati_GAS = takeData(file_dati_GAS)

# Tempo
time = np.arange(0, 300, 1)
tmax = len(dati_SOLIDO)

# Calcolo della varianza per ciascun sistema
var_SOLIDO = np.sum((dati_SOLIDO - np.mean(dati_SOLIDO))**2) / len(dati_SOLIDO)
var_LIQUIDO = np.sum((dati_LIQUIDO - np.mean(dati_LIQUIDO))**2) / len(dati_LIQUIDO)
var_GAS = np.sum((dati_GAS - np.mean(dati_GAS))**2) / len(dati_GAS)

# Inizializzazione dell'autocorrelazione
autocorr_SOLIDO = np.zeros(len(time))
autocorr_LIQUIDO = np.zeros(len(time))
autocorr_GAS = np.zeros(len(time))

# Calcolo dell'autocorrelazione
for j, t in enumerate(time):
    ac_SOLIDO = np.sum(dati_SOLIDO[:tmax-t]*dati_SOLIDO[t:]) / (tmax - t)
    ac_SOLIDO -= np.sum(dati_SOLIDO[:tmax-t]) * np.sum(dati_SOLIDO[t:]) / (tmax - t)**2
    autocorr_SOLIDO[j] = ac_SOLIDO / var_SOLIDO

    ac_LIQUIDO = np.sum(dati_LIQUIDO[:tmax-t]*dati_LIQUIDO[t:]) / (tmax - t)
    ac_LIQUIDO -= np.sum(dati_LIQUIDO[:tmax-t]) * np.sum(dati_LIQUIDO[t:]) / (tmax - t)**2
    autocorr_LIQUIDO[j] = ac_LIQUIDO / var_LIQUIDO

    ac_GAS = np.sum(dati_GAS[:tmax-t]*dati_GAS[t:]) / (tmax - t)
    ac_GAS -= np.sum(dati_GAS[:tmax-t]) * np.sum(dati_GAS[t:]) / (tmax - t)**2
    autocorr_GAS[j] = ac_GAS / var_GAS

# Trova il punto in cui l'autocorrelazione cade sotto una soglia
threshold = 0.0005  
lag_SOLIDO = np.where(autocorr_SOLIDO < threshold)[0][0]
lag_LIQUIDO = np.where(autocorr_LIQUIDO < threshold)[0][0]
lag_GAS = np.where(autocorr_GAS < threshold)[0][0]

# Stima della larghezza dei blocchi
block_size_SOLIDO = lag_SOLIDO + 1
block_size_LIQUIDO = lag_LIQUIDO + 1
block_size_GAS = lag_GAS + 1

# Plot delle autocorrelazioni
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(time, autocorr_SOLIDO)
plt.axvline(x=lag_SOLIDO, color='red', linestyle='--', label=f'Soglia {lag_SOLIDO}')
plt.xlabel('Tempo')
plt.ylabel('Autocorrelazione')
plt.title('Autocorrelazione per $U/N$  (sistema allo stato solido)')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(time, autocorr_LIQUIDO)
plt.axvline(x=lag_LIQUIDO, color='red', linestyle='--', label=f'Soglia {lag_LIQUIDO}')
plt.xlabel('Tempo')
plt.ylabel('Autocorrelazione')
plt.title('Autocorrelazione per $U/N$ (sistema allo stato liquido)')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(time, autocorr_GAS)
plt.axvline(x=lag_GAS, color='red', linestyle='--', label=f'Soglia {lag_GAS}')
plt.xlabel('Tempo')
plt.ylabel('Autocorrelazione')
plt.title('Autocorrelazione per $U/N$  (sistema allo stato aeriforme)')
plt.legend()

plt.tight_layout()
plt.show()
```

_La soglia si riferisce al numero minimo di blocchi per avere una autocorrelazione minore di 5 parti per 10000_

```python

#PUNTO SECONDO

def data_blocking(dati):
 
    # Calculate the mean of each block directly without splitting
    block_means = np.mean(dati, axis=1) # Calculate the mean of each sub-array
    
    # Calculate the global average and global average squared
    glob_av = np.mean(block_means)
    glob_av2 = np.mean(block_means**2)
    
    # Calculate the error
    blocks = len(dati)
    error = np.sqrt((glob_av2 - glob_av**2) / (blocks - 1))
    return error

# Preparazione per i grafici
plt.figure(figsize=(17, 15))

# Griglia 3x1 per i grafici
grid = plt.GridSpec(3, 1, wspace=0.3, hspace=0.5)

# Grafico 1: stato solido
plt.subplot(grid[0, 0])
L = np.arange(10, 1000, 1)
err = np.zeros(len(L))

for i, Len in enumerate(L):
    N = len(dati_SOLIDO) // Len
    dati_reshaped = dati_SOLIDO[:N*Len].reshape((N, Len))
    err[i] = data_blocking(dati_reshaped)

plt.plot(L, err)
plt.title('Andamento dell incertezza di U/N    (sistema allo stato solido)')
plt.xlabel('Dimensione dei blocchi')
plt.ylabel('Errore')
plt.grid(True)
plt.yscale("log")

# Grafico 2: stato liquido
plt.subplot(grid[1, 0])
L = np.arange(10, 1000, 1)
err = np.zeros(len(L))

for i, Len in enumerate(L):
    N = len(dati_LIQUIDO) // Len
    dati_reshaped = dati_LIQUIDO[:N*Len].reshape((N, Len))
    err[i] = data_blocking(dati_reshaped)

plt.plot(L, err)
plt.title('Andamento dell incertezza di U/N   (sistema allo stato liquido)')
plt.xlabel('Dimensione dei blocchi')
plt.ylabel('Errore')
plt.grid(True)
plt.yscale("log")

# Grafico 3: stato aeriforme
plt.subplot(grid[2, 0])
L = np.arange(10, 1000, 1)
err = np.zeros(len(L))

for i, Len in enumerate(L):
    N = len(dati_GAS) // Len
    dati_reshaped = dati_GAS[:N*Len].reshape((N, Len))
    err[i] = data_blocking(dati_reshaped)

plt.plot(L, err)
plt.title('Andamento dell incertezza di U/N    (sistema allo stato aeriforme)')
plt.xlabel('Dimensione dei blocchi')
plt.ylabel('Errore')
plt.grid(True)
plt.yscale("log")

plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_path):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(os.path.join(os.getcwd(), file_path), 'r') as file:
            # Salta la prima riga
            next(file)
            # Itera su ogni riga del file
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore
                dati_riga = riga.strip().split()
                
                # Converte i primi tre dati in float e aggiunge alla lista dati
                dati_float = [float(dato) for dato in dati_riga[:3]]
                dati.append(dati_float)
    except FileNotFoundError:
        print(f"Errore: il file {file_path} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy

directory = str(os.getcwd())

# Percorsi ai file dei dati
file_dati_SOLIDO_MC = os.path.join(directory, "Monte_Carlo/OUTPUT_solido/gofr.dat")
file_dati_LIQUIDO_MC = os.path.join(directory, "Monte_Carlo/OUTPUT_liquido/gofr.dat")
file_dati_GAS_MC = os.path.join(directory, "Monte_Carlo/OUTPUT_gas/gofr.dat")

file_dati_SOLIDO_DM = os.path.join(directory, "Dinamica_Molecolare/OUTPUT_solido/gofr.dat")
file_dati_LIQUIDO_DM = os.path.join(directory, "Dinamica_Molecolare/OUTPUT_liquido/gofr.dat")
file_dati_GAS_DM = os.path.join(directory, "Dinamica_Molecolare/OUTPUT_gas/gofr.dat")

# Lettura dati Monte Carlo
dati_SOLIDO_MC = takeData(file_dati_SOLIDO_MC)
x_values_SOLIDO_MC = dati_SOLIDO_MC[:, 0]
y_values_SOLIDO_MC = dati_SOLIDO_MC[:, 1]
y_errors_SOLIDO_MC = dati_SOLIDO_MC[:, 2]

dati_LIQUIDO_MC = takeData(file_dati_LIQUIDO_MC)
x_values_LIQUIDO_MC = dati_LIQUIDO_MC[:, 0]
y_values_LIQUIDO_MC = dati_LIQUIDO_MC[:, 1]
y_errors_LIQUIDO_MC = dati_LIQUIDO_MC[:, 2]

dati_GAS_MC = takeData(file_dati_GAS_MC)
x_values_GAS_MC = dati_GAS_MC[:, 0]
y_values_GAS_MC = dati_GAS_MC[:, 1]
y_errors_GAS_MC = dati_GAS_MC[:, 2]

# Lettura dati Dinamica Molecolare
dati_SOLIDO_DM = takeData(file_dati_SOLIDO_DM)
x_values_SOLIDO_DM = dati_SOLIDO_DM[:, 0]
y_values_SOLIDO_DM = dati_SOLIDO_DM[:, 1]
y_errors_SOLIDO_DM = dati_SOLIDO_DM[:, 2]

dati_LIQUIDO_DM = takeData(file_dati_LIQUIDO_DM)
x_values_LIQUIDO_DM = dati_LIQUIDO_DM[:, 0]
y_values_LIQUIDO_DM = dati_LIQUIDO_DM[:, 1]
y_errors_LIQUIDO_DM = dati_LIQUIDO_DM[:, 2]

dati_GAS_DM = takeData(file_dati_GAS_DM)
x_values_GAS_DM = dati_GAS_DM[:, 0]
y_values_GAS_DM = dati_GAS_DM[:, 1]
y_errors_GAS_DM = dati_GAS_DM[:, 2]

# Crea i subplot affiancati
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
plt.subplots_adjust(hspace=0.5)

# Grafico stato solido
axs[0].errorbar(x_values_SOLIDO_MC, y_values_SOLIDO_MC, yerr=y_errors_SOLIDO_MC, label='Monte Carlo')
axs[0].errorbar(x_values_SOLIDO_DM, y_values_SOLIDO_DM, yerr=y_errors_SOLIDO_DM, label='Dinamica Molecolare')
axs[0].set_xlabel('Distanza')
axs[0].set_ylabel('g(r)')
axs[0].set_title('g(r) stato solido')
axs[0].grid(True)
axs[0].legend()

# Grafico stato liquido
axs[1].errorbar(x_values_LIQUIDO_MC, y_values_LIQUIDO_MC, yerr=y_errors_LIQUIDO_MC, label='Monte Carlo')
axs[1].errorbar(x_values_LIQUIDO_DM, y_values_LIQUIDO_DM, yerr=y_errors_LIQUIDO_DM, label='Dinamica Molecolare')
axs[1].set_xlabel('Distanza')
axs[1].set_ylabel('g(r)')
axs[1].set_title('g(r) stato liquido')
axs[1].grid(True)
axs[1].legend()

# Grafico stato aeriforme
axs[2].errorbar(x_values_GAS_MC, y_values_GAS_MC, yerr=y_errors_GAS_MC, label='Monte Carlo')
axs[2].errorbar(x_values_GAS_DM, y_values_GAS_DM, yerr=y_errors_GAS_DM, label='Dinamica Molecolare')
axs[2].set_xlabel('Distanza')
axs[2].set_ylabel('g(r)')
axs[2].set_title('g(r) stato aeriforme')
axs[2].grid(True)
axs[2].legend()

plt.show()
```
