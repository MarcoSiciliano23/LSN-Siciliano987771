In this exercise we use genetic algorithms to solve numerically the traveling salesman problem, which consists in minimizing the length of a closed path that connects $n$ fixed points and that visits any point once and only once.

## Genetic algorithms
<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
By genetic algorithm we mean a technique that can be used to solve a large class of optimization problems, by a process that I would call "trial and error" that is inspired by natural evolution.  
As just as in real environment, we simulate a population of individuals, fully characterized by a series of adjustable parameters that determine the individual's performance in terms of the cost function it is our goal to optimize. Every generation of individuals is subject to a process of selection, which determines its probability to carry its genome to the next generation.  
Since it's very unlikely (read: practically impossible) that the values of the parameters that optimize the cost function are already present in the population since the first generation, genetic variability must be included. This is achieved with the introduction of mutation operators that change in a random way the genes of the selected parents before carrying them to the next generation, and of crossing operators, that select two different individuals and exchanges part of their genome before carrying it to the offspring.  
It is important to note that all the aforementioned operators must generate offspring that still respect the bonds given by the problem. The process of genetic algorithms can be summarized in the following way: we start with a random population, and select the parents of the next generation according to their fitness. Then we apply crossing to each pair of parents and finally mutate the offspring, and repeat.

## The traveling salesman problem
<img src="Samsa.jpg" width="600">

This problem poses the question to find, given $n$ cities $\{ \vec{c_0}, ... , \vec{c}_{n-1} \}$ and the distance between any pair of cities, the shortest closed path that connects all the cities.  
We can use a genetic algorithm to solve this problem. The cost function to optimize is (we choose to use the $L^{(2)}$ metric):
$$ L^{(2)}(\{ \vec{c_1}, ... , \vec{c_n} \}) = \sum_{i=0}^{n-1} |\vec{c_i} - \vec{c}_{i+1}|^2 \text{.}$$
Where the ciclicity of the path is expressed with the periodic boundary condition $\vec{c}_{n} = \vec{c_0}$.  
An individual is defined as a permutation of the indeces of cities which represent the order in which the salesman will visit them:
$$ [\vec{c}_{\sigma(0)},..., \vec{c}_{\sigma(n-1)}] \text{,}$$
where $\sigma $ is a generic permutation of the vector of indeces. We are looking for the $\sigma_{best}$ that minimizes the cost function $L^{(2)}$.  
We start off with a population of randomly picked permutations.

### Selection operator
After being generated, the individuals are ordered according to their fitness (value of $L^{(2)}$), from lowest to highest. The selection operator chooses an individual according to the value of the random variable:
$$ p = \left \lfloor{M r^2} \right \rfloor \text{,} $$
where $\rho(r)=\chi_{[0, 1)}$ and $M$ is the number of individuals in the population. It is immediate to prove that the probability distribution of $p$ is proportional to $ \frac{1}{\sqrt{p}} $, thus it makes it more likely to select low indeces instead of high ones.

### Crossing operator
One fundamental aspect of genetic algorithms is crossover, in which two parent individuals exchange part of their genome with probability $p_c$ and pass it to the offspring, which is therefore composed of two individuals with a completely new genome.  
Since if we naively exchanged the indeces of the two parents from a certain index in the genome we could not satisfy the bonds, since the offspring could have repetitions in their genome, we implement the crossing in the following way:
We consider a string of indeces of each parent (e.g. $[0, 1 | 2, 3, 4]$ and $[0, 3 | 4, 1, 2]$) and reorder them such that they appear in the order the other indeces appear in the other parent. (In our example, the two genomes become $[0, 1 | 4, 2, 3]$ and $[0, 3 | 1, 2, 4]$).

### Mutation operators
The next generation of individuals should be different from their parents in order to explore ergodically the space of permutations. That's why we define mutation operators that act on the selected individuals with a probability $p_m$. We include four different mutations that always leave the first element of the sequence of cities unchanged, in order to reduce the probability of different permutations that actually describe the same path to compete with each other, slowing down the process of optimization.
1. Point-like permutation of two indeces: e.g. $ [0, 1, 2, 3, 4] \rightarrow [0, 1 , 3, 2, 4]$.
2. Permutation of portions of length $m$: e.g. $ [0, 1, 2, 3, 4] \rightarrow [0, 3, 4, 1, 2]$.
3. Shift by $n$ of a portion of length $m$: e.g. $ [0, 1, 2, 3, 4] \rightarrow  [0, 1, 4, 2, 3]$.
4. Inversion of a portion of length $m$: e.g. $ [0, 1, 2, 3, 4] \rightarrow  [0, 1, 4, 3, 2]$.

## Parameters of simulation
We solve the TSP in two cases: in the first, all the $n=34$ cities are disposed on a circumference, and in the second they are uniformly distributed inside a square.
For both the cases we use a populaton of $M=500$ individuals carried on for $N=1000$ generations.
For the problem on a circumference, we use $p_c = 95\%$ and $p_m = 5\%$, while for the problem in a square we use $p_c = 95\%$ and $p_m = 11\%$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#best fitness as a function of the number of generations
n_gens_circle, best_circle = np.loadtxt("OUTPUT/best_fitness_circle.dat", usecols=(0, 1), skiprows=1, unpack=True)
n_gens_square, best_square = np.loadtxt("OUTPUT/best_fitness_square.dat", usecols=(0, 1), skiprows=1, unpack=True)

#average fitness as a function of the number of generations
n_gens_circle, average_circle = np.loadtxt("OUTPUT/fitness_circle.dat", usecols=(0, 1), skiprows=1, unpack=True)
n_gens_square, average_square = np.loadtxt("OUTPUT/fitness_square.dat", usecols=(0, 1), skiprows=1, unpack=True)
```

```python
# Create a figure with two subplots arranged in a single row
fig, axx = plt.subplots(1, 2, figsize=(12, 6), tight_layout=True)

# Plot the best fitness per generation for the circle data on the first subplot
axx[0].plot(n_gens_circle, best_circle, linewidth=0.8, marker=".")
axx[0].set_xlabel("generation")  # Set the x-axis label for the first subplot
axx[0].set_ylabel(r"$L^{(2)}$")  # Set the y-axis label for the first subplot with LaTeX formatting
axx[0].set_title("Best fitness per generation on circle")  # Set the title for the first subplot
axx[0].grid(True)  # Enable grid for the first subplot

# Plot the best fitness per generation for the square data on the second subplot
axx[1].plot(n_gens_square, best_square, linewidth=0.8, marker=".")
axx[1].set_xlabel("generation")  # Set the x-axis label for the second subplot
axx[1].set_ylabel(r"$L^{(2)}$")  # Set the y-axis label for the second subplot with LaTeX formatting
axx[1].set_title("Best fitness per generation in square")  # Set the title for the second subplot
axx[1].grid(True)  # Enable grid for the second subplot
```

```python
# Create a figure with two subplots arranged in a single row
fig, axx = plt.subplots(1, 2, figsize=(12, 6), tight_layout=True)

# Plot the average fitness per generation for the circle data on the first subplot
axx[0].plot(n_gens_circle, average_circle, linewidth=0.8, marker=".")
axx[0].set_xlabel("generation")  # Set the x-axis label for the first subplot
axx[0].set_ylabel(r"$\langle L^{(2)} \rangle$")  # Set the y-axis label for the first subplot with LaTeX formatting
axx[0].set_title("Average fitness per generation on circle")  # Set the title for the first subplot
axx[0].grid(True)  # Enable grid for the first subplot

# Plot the average fitness per generation for the square data on the second subplot
axx[1].plot(n_gens_square, average_square, linewidth=0.8, marker=".")
axx[1].set_xlabel("generation")  # Set the x-axis label for the second subplot
axx[1].set_ylabel(r"$\langle L^{(2)} \rangle$")  # Set the y-axis label for the second subplot with LaTeX formatting
axx[1].set_title("Average fitness per generation in square")  # Set the title for the second subplot
axx[1].grid(True)  # Enable grid for the second subplot
```

Note that the best fitness value at the $n^{th}$ generation tends to become the average fitness value (for the best half of the population) at the ${(n+1)}^{th}$ generation, proving that the best genes are actually propagating in the population.

```python
#load cities
cities_circle = np.loadtxt("OUTPUT/cities_circle.dat", usecols=(1, 2), skiprows=1)
cities_circle = np.transpose(cities_circle)

cities_square = np.loadtxt("OUTPUT/cities_square.dat", usecols=(1, 2), skiprows=1)
cities_square = np.transpose(cities_square)
```

```python
# Load the best genes (city indices) for circumference and square from files
genes_circ = np.loadtxt("OUTPUT/genes_circle.dat", usecols=0, skiprows=1)
genes_square = np.loadtxt("OUTPUT/genes_square.dat", usecols=0, skiprows=1)

# Extract x and y coordinates of cities on the circumference and square
circle_x = cities_circle[0]
circle_y = cities_circle[1]
square_x = cities_square[0]
square_y = cities_square[1]

# Initialize lists to store the path coordinates for the circle and square
path_circle_x = [0 for i in range(circle_x.shape[0])]
path_circle_y = [0 for i in range(circle_y.shape[0])]
path_square_x = [0 for i in range(square_x.shape[0])]
path_square_y = [0 for i in range(square_y.shape[0])]

# Loop through each gene (city index) to create the path coordinates
for i in range(genes_circ.shape[0]):
    # Map the genes to the corresponding city coordinates for the circle
    path_circle_x[i] = cities_circle[0][int(genes_circ[i])]
    path_circle_y[i] = cities_circle[1][int(genes_circ[i])]
    
    # Map the genes to the corresponding city coordinates for the square
    path_square_x[i] = cities_square[0][int(genes_square[i])]
    path_square_y[i] = cities_square[1][int(genes_square[i])]
```

```python
# Create a figure with two subplots arranged in a single row
fig, axx = plt.subplots(1, 2, figsize=(12, 6), tight_layout=True)

# Plot the best path for cities on the circumference on the first subplot
axx[0].plot(path_circle_x, path_circle_y, linewidth=0.8, marker=".")
axx[0].set_xlabel("x")  # Set the x-axis label for the first subplot
axx[0].set_ylabel("y")  # Set the y-axis label for the first subplot
axx[0].set_title("Best path for cities on circumference")  # Set the title for the first subplot
axx[0].grid(True)  # Enable grid for the first subplot

# Plot the best path for cities in the square on the second subplot
axx[1].plot(path_square_x, path_square_y, linewidth=0.8, marker=".")
axx[1].set_xlabel("x")  # Set the x-axis label for the second subplot
axx[1].set_ylabel("y")  # Set the y-axis label for the second subplot
axx[1].set_title("Best path for cities in square")  # Set the title for the second subplot
axx[1].grid(True)  # Enable grid for the second subplot
```
