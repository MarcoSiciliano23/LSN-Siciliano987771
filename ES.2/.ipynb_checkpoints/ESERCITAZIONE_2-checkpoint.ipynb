#### ESERCIZIO 2.1
Alla classe _Random_ sono stati aggiunti due metodi:  
- _IntegraMedia_ che calcola l'integrale di una funzione, sfruttando numeri generati uniformemente tra 0 e il massimo della funzione;
- _AR_ che implementa la tecnica di _Accept-Reject_ per generare numeri secondo una qualunque distribuzione (implementata attraverso una classe concreta, che eredita dalla classe astratta _FunzioneBase_).


Per calcolare l'integrale $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ sono stata eseguite $10^4$ simulazioni: dapprima con il metodo _IntegraMedia_, successivamente con l'_importance sampling_ (sfruttando _AR_).  
In intrambi i casi, le incertezze sono state calcolate con il metodo della _media a blocchi_.
<br><br>




La distribuzione di probabilità utilizzata per l'_importance sampling_ è  $ \ p(x) = -\frac{3}{2} x^2 +\frac{3}{2} $


```python
import numpy as np
import matplotlib.pyplot as plt

def takeData(file_nome):
    dati = []  # Lista per contenere i dati letti
    try:
        with open(file_nome, 'r') as file:
            for riga in file:
                # Separa i dati utilizzando uno spazio come delimitatore (puoi modificare in base al formato)
                dati_riga = riga.strip().split()
                
                # Converte i dati in float e aggiunge alla lista dati
                dati_float = [float(dato) for dato in dati_riga]
                dati.append(dati_float)
    except FileNotFoundError:
        print(f"Errore: il file {file_nome} non esiste.")
    except Exception as e:
        print(f"Errore durante la lettura del file: {e}")

    return np.array(dati)  # Converte la lista di liste in un array NumPy

# Primo grafico
file_dati1 = "unif_sampl.out"
dati1 = takeData(file_dati1)
x_values1 = dati1[:, 0]  # Valori
y_errors1 = dati1[:, 1]   # Errori
             
N = 100 # Numero di blocchi

x = np.arange(0, N, 1)

# Crea i subplot affiancati
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Primo grafico
ax1.errorbar(x, x_values1, yerr=y_errors1)
ax1.set_xlabel('#Blocchi')
ax1.set_ylabel('<I>')
ax1.set_title("Andamento di <I> con uniform samplig")
ax1.grid(True)

# Secondo grafico
ax2.errorbar(x, x_values2, yerr=y_errors2)
ax2.set_xlabel('#Blochhi')
ax2.set_ylabel('<I>')
ax2.set_title('Andamento di <I> con importance samplig')
ax2.grid(True)

plt.tight_layout()
plt.show()

```

#### ESERCIZIO 2.2
Alla classe _Random_ è stato aggiunto il metodo _Rnd_Walk_ (in _overloading_):   
- _vector<vector<\int>> Rnd_Walk (int passi, int a)_ per implementare un _random walk_ discreto;
- _vector<vector<\double>> Rnd_Walk (int passi, double b)_ per implementare un _random walk_ continuo.


<br>
Nel caso discreto, la probabilità di muoversi in una direzione è data dall'estrazione casuale di un numero intero tra 0 e 6:  
  ogni numero indica un spostamento, avanti(_+a_) o indietro(_-a_), in ciascuna direzione {0,1}  $ \ \hat{x} \ $,  {2,3}  $ \ \hat{y} \ $,    {4,5}$ \ \hat{z}$.

Nel caso continuo, la probabilità di muoversi in una direzione è data dall'estrazione casuale dei valori per angolo azimutale $\phi \ $ e l'angolo polare $\theta$:
  - $\phi \ $  estratto uniformemente tra 0 e 2$\pi$;
  - $\theta \ $  estrato attraverso l'inversione della _cumulativa<sup>1</sup>_  tra 0 e $\pi$.

<br>

Per calcolare $ \ \sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}} \ $ in funzione dei _passi_ $ \ i\in [0,10^2]$ è stato usato il metodo della _media a blocchi_ per $\ \langle |\vec{r}_N|^2 \rangle_{RW} \ $ a ogni passo _i_; ai fini del grafico invece, è stata estratta la radice solo dell'ultimo valore di ogni media (propagando l'errore).  
<br>
L'andamento di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}\ $ , per ambedue i _random walk_, segue $f(N)=k\sqrt{N} \ $ con $k=1$, indicando un comportamento diffusivo.

  <br><br>


  

<sup>1)</sub><sub><sup>$ F^{-1} = arccos(1 - 2 u) \ $, dove $u \in [0,1) $ è uniforme. </sup></sub>

```python
import math
# Dati per il primo grafico
file_dati_discreto = "walk_discreto.out"
dati_discreto = takeData(file_dati_discreto)
x_values_discreto = dati_discreto[:, 0]  # Valori
y_errors_discreto = dati_discreto[:, 1]  # Errori

# Dati per il secondo grafico
file_dati_continuo = "walk_continuo.out"
dati_continuo = takeData(file_dati_continuo)
x_values_continuo = dati_continuo[:, 0]  # Valori
y_errors_continuo = dati_continuo[:, 1]  # Errori

# Parametri comuni
M = 100  # Numero totale di passi
L = 1
x = np.arange(0, M, L)

# Amplifica gli errori per la visualizzazione
y_errors_amplificati_discreto = y_errors_discreto * 5
y_errors_amplificati_continuo = y_errors_continuo * 5

# Crea i subplot affiancati
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Primo grafico
ax1.errorbar(x, x_values_discreto, yerr=y_errors_amplificati_discreto)
ax1.plot(x, np.sqrt(x), 'r-', label= '√N')  # Funzione sqrt(N) in rosso
ax1.set_xlabel('#passi')
ax1.set_ylabel('√<r^2>')
ax1.set_title('Random walk discreto\nGrafico con errori amplificati di un fattore 5')
ax1.grid(True)
ax1.legend()


# Secondo grafico
ax2.errorbar(x, x_values_continuo, yerr=y_errors_amplificati_continuo)
ax2.plot(x, np.sqrt(x), 'r-', label= '√N')  # Funzione sqrt(N) in rosso
ax2.set_xlabel('#passi')
ax2.set_ylabel('√<r^2>')
ax2.set_title('Random walk continuo\nGrafico con errori amplificati di un fattore 5')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
